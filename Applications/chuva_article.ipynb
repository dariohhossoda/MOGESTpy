```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rcParams
from matplotlib import rc
from datetime import datetime, timedelta

rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['CMU Serif']
rcParams['text.usetex'] = True
```

```python
df = pd.read_excel('smap_jundiai_chuva.xlsx')
```

```python
fig, ax = plt.subplots(figsize=(10/2.54, 10/2.54), constrained_layout=False)

x = [i + 1 for i in range(365)]

bar_width = 2

ax.bar(x, df['P'], label='$P$ - Sim.', linewidth=2.5, color='black', width=bar_width)
ax.bar(x, df['P4d'], label='$P$ - $\Delta t = 4$ days', linewidth=1.5, width=bar_width)
ax.bar(x, df['P8d'], label='$P$ - $\Delta t = 8$ days', linewidth=1.5, width=bar_width)
ax.bar(x, df['P15d'], label='$P$ - $\Delta t = 15$ days', linewidth=1.5, width=bar_width)

ax2 = ax.twinx()

ax2.plot(x, df['P'].cumsum(), label='P Sim.', linewidth=0.5, color='black')
ax2.plot(x, df['P4d'].cumsum(), label='P - $\Delta t = 4$ days', linewidth=0.5)
ax2.plot(x, df['P8d'].cumsum(), label='P - $\Delta t = 8$ days', linewidth=0.5)
ax2.plot(x, df['P15d'].cumsum(), label='P - $\Delta t = 15$ days', linewidth=0.5)

ax.set_xlabel('Time (days)')
ax.set_ylabel('Precipitation (mm)')
ax2.set_ylabel('Precipitação Cumulative Sum (mm)')

ax.legend(ncol=2, loc=(0, 1.))





ax.set_xlim(0, 365)
ax.set_ylim(0)
# plt.savefig('H:/GitHub/Thesis/images/analise/chuvas1.pdf')
plt.show()
```
