# Exemplo de aplicação do Modelo 0D

Para começar a utilizar a aplicação, é necessário chamar o módulo de qualidade e o modelo

```python
import os
import matplotlib.pyplot as plt
from matplotlib import rcParams

# É necessário tornar a pasta do repositório como caminho ativo
os.chdir('H:/GitHub/MOGESTpy')
from ZeroD import ZeroDimensional as zd
```

A seguir, as variáveis de entrada serão inicializadas. Para mais informações, consulte a [documentação](../README.md)

```python
vector_size = 10
Qin = [50 for _ in range(vector_size)]
Qout = [50 for _ in range(vector_size)]
Cin = [60 for _ in range(vector_size)]
V = [1000 for _ in range(vector_size)]
As = [50 for _ in range(vector_size)]
v = [.5 for _ in range(vector_size)]
k = [.05 for _ in range(vector_size)]

dt = 50
```

Para o exemplo, será criado uma variável representado um reator (reactor) em que zd é a classe importada

```python
reactor = zd(V, Qin, Qout, Cin, As, k, v, dt)
```

Agora é só fazer com que o modelo seja executado através da seguinte linha

```python
reactor.RunModel()
```

Os resultados podem ser verificados da seguinte forma:

```python
result = reactor.Output.ConcentrationOut
print(result)
```

Dessa forma é possível ver a evolução da concentração em função do tempo no reator como:

```python
# Configurações de gráfico
rcParams['font.family'] = 'serif'
rcParams['font.serif'] = ['Times New Roman']

x_axis = [i for i in range(vector_size)]

plt.legend('Concentração')
plt.title('Concentração vs Tempo')
plt.xlabel('Tempo (s)')
plt.ylabel('Concentração (mg/L)')
plt.grid(True)

plt.plot(x_axis, result)
```
